```python
def naivelog(x):
  n=15
  y = x
  for i in range(n):
    y = math.sqrt(y)
  l = y-1
  for i in range(n):
    l = l*2
  return l
```

1. Initialization: Set n to 15, and initialize y as x.
    
2. First Loop (Square Root): Iterate over a loop n times. In each iteration, take the square root of y. This operation is repeated 15 times.

3. Update l: Set l to the final value of y minus 1.
    
3. Second Loop (Multiply by 2): Iterate over a loop n times. In each iteration, multiply l by 2.
    
4. Return l: The final value of l is returned.

The algorithm naivelog(x), repeatedly applies the square root operation to the input y=x a total of n times, gradually reducing it until ln(y) ≈ y-1 (because y ≈ 1). Then, l=y-1 is multiplied by 2, n times. Finally, the result is returned.

This technique of computation of logarithm is known as argument halving, it allows us to compute values of ln(x) for large x relatively precisely. In this particular implementation of naivelog, n=15. 

$ ln(y) -> ln(y_2) -> ... -> ln(y_n) ≈ y_n - 1 (y_n ≈ 1)$ This is the goal to reach, because ln(1)=0, so very accurate.

(PS: I couldnt find a way to make a "pretty" square root symbol.)

The parameter n in the naivelog(x) algorithm represents the number of iterations in the iterative processes. Specifically, it dictates how many times the algorithm applies the square root operation in the initial reduction phase and how many times it performs the multiplication in the subsequent refinement phase. The value of n influences the precision and accuracy of the algorithm's approximation of the ln(x), with larger values of n generally leading to more accurate results but also potentially requiring more computational resources.

When x=0, the variable y remains 0 throughout the successive square root operations. Consequently, the value of l=y-1 is set to be -1. This resultant value of l=-1 is multiplied by 2 n times, here n=15. Then, returned. Which is wrong, since ln(0) is undefined.

When x=1, the variable y remains 1 throughout the successive square root operations. Consequently, the value of l=y-1 is set to be 0. This resultant value of l=0 is multiplied by 2 n times, here n=15, results in l=0. Thus, it will return l=0, which is the right answer.

### Analysis

The functions naivelog and mylog behave differently because they don't use the same argument reduction method, which impacts the errors and the speed of convergence of the approximations made by these two algorithms.
- mylog, the argument is reduced a little bit more than by half at each iteration.
- naivelog, the argument is reduced by square at each iteration.

Convergence rate:
- For mylog, as observed above, the combination of division and square roots facilitates faster convergences.
- For naivelog, the convergence is relatively slow, and the convergence reach's a maximum point, because by increasing n after n=30, we are loosing accuracy for naivelog.

The increase error for naivelog as n becomes larger than 30 is likely due to numerical instability, the limitations of floating-point arithmetic and cancellation of digits.

1. Numerical Instability:
- As n increases, naivelog performs more iterations of square root operations.
- Repeated square roots on a number close to 1 can introduce instability.
- Numerical instability leads to a loss of precision, and errors accumulate over iterations.
2. Rounding Error:
- Floating-point arithmetic has limitations in representing real numbers with infinite precision.
- The result of each square root operation is subject to rounding errors.
- As n increases, the accumulation of rounding errors becomes more pronouced.
3. Cancellation Error:
- In floating-point arithmetic, subtracting two similar numbers can result in cancellation of digits (error).
- A large number of square roots will give a result that is very close to 1, subtracting it from 1 in the next step (l=y-1) may lead to significant digits cancellation. 

The mylog algorithm doesnt suffer from cancellation error as it doesn't have any subtraction operation past the first step of the algorithm. Thus, it still suffers from rounding error, but there is no possibility for cancellation error. Thus, this is why mylog keeps getting more accurate as n grows.

## Exponential

```python
def myexp(x, n):
    y = x
    for i in range(n):
        y = y/2
    l = 1 + y
    for i in range(n):
        l = l **2
        
    return l

# Calculate errors for various values of n
values = list(range(1, 15))
n_values = [15, 20, 25, 30, 35]

for n in n_values:
    print(f"\nFor n = {n}:")
    for value in values:
        true_result = math.exp(value)
        myexp_result = myexp(value, n)
        error_myexp = abs(myexp_result - true_result)
        relative_error = error_myexp / abs(true_result)
        print(f"Value: {value}, Absolute Error (myexp): {error_myexp}, Relative Error (myexp): {relative_error}")
```

The idea for my implementation is using:

$e^x ≈ (1+y)^N = (1 + x/N)^N$ where $y = x/N, N = 2^k$

## Sine

I will use the small angle approximation.
$sin(x) ≈ x$ for very small x.

(Source: https://en.wikipedia.org/wiki/Small-angle_approximation) 

And this argument reduction technique from the notebook:


```python
def mysin(x, n):
    y = x
    for i in range(n):
        y = y/2
        
    for i in range(n):
        y = 2*y*math.sqrt(1 - y**2)
        
    return y

# Calculate errors for various values of n
values = [0.1 * i for i in range(18)]
n_values = [15, 20, 25, 30, 35]

for n in n_values:
    print(f"\nFor n = {n}:")
    for value in values:
        true_result = math.sin(value)
        mysin_result = mysin(value, n)
        error_mysin = abs(mysin_result - true_result)
        
        # Check if true_result is not zero before calculating relative error
        if true_result != 0:
            relative_error = error_mysin / abs(true_result)
        else:
            relative_error = float('inf')  # Set relative error to infinity for zero true_result

        print(f"Value: {value}, Absolute Error (mysin): {error_mysin}, Relative Error (mysin): {relative_error}")
```

I will implement a CORDIC algorithm that computes tan by using vector rotations. Which will allow me to estimate: $tan(z) = x/y$. The angle z will be decomposed into a sum of angles, each corresponding to 0 or $arctan(2^{-n})$. Then the function iteratively rotates a vector (x,y) to approximate the original angle z. 

(PS: I chose 2 instead of 10, because I kept trying the exact same code with 10 and it wasnt giving good relative errors.)


```python
def myarctan(x):
    y = x
    n=30
    
    for i in range(n):
        y = y/(1+math.sqrt(1+y**2))
        
    for i in range(n):
        y = 2*y
        
    return y

arctan_table = [myarctan(2 ** -i) for i in range(0, 17)]

def cordic_tan(z):
    binary_angles = []
    
     # Break down the angle into sum of angles, where each angle is 0 or atan(2**-i)
    for i in range(1, 17):
        if z >= arctan_table[i]:
            z -= arctan_table[i]
            binary_angles.append(1)
        else:
            binary_angles.append(0)

     # Approximate tan of the remaining angle (very small)
    tan_remaining = (3 * z) / (3 - z * z)
    
    #Initial values for vector components
    y = tan_remaining
    x = 1
    
    #Rotate the vector for each angle in the original sum of angles
    for i in range(16, 0, -1):
        if binary_angles[i - 1]:
            y_tmp = y + x * 2**-i
            x = x - y * 2**-i
            y = y_tmp

    return y / x


# Example usage
test_values = [0.1 * i for i in range(18)]
for value in test_values:
    true_result = math.tan(value)
    cordic_result = cordic_tan(value)
    error_cordic = abs(cordic_result - true_result)

    # Check if true_result is not zero before calculating relative error
    if true_result != 0:
        relative_error = error_cordic / abs(true_result)
    else:
        relative_error = error_cordic

    print(f"Input: {value}, True Tan: {true_result}, CORDIC Tan: {cordic_result}, Relative Error: {relative_error}")
```

As we can see, my CORDIC implementation of tan gives very little relative error, thus it is good.

From textbook:
    
    

```python
def cordic_atan(A):
    # Initialize variables
    if A == 0:
        return 0
    
    Yt = abs(A)
    Xt = 1.0
    angle = 0.0
    sign_A = 1 if A >= 0 else -1

    # Rotation process
    for i in range(1, 17):
        factor = 1 / (1 << i)
        if Yt >= Xt * myarctan(2 ** -i):
            Yt, Xt, angle = Yt - Xt * factor, Xt + Yt * factor, angle + myarctan(2 ** -i)

    # Clockwise rotation until Yt becomes negative
    while Yt >= 0:
        Yt, Xt, angle = Yt - Xt, Xt + Yt, angle - myarctan(2 ** -16)

    # Apply sign to determine correct quadrant
    return sign_A * angle

# Example usage
test_values = [0.1 * i for i in range(11)]
for value in test_values:
    true_result = math.atan(value)
    cordic_result = cordic_atan(value)
    error_cordic = abs(cordic_result - true_result)

    print(f"Input: {value}, True arctan: {true_result}, CORDIC arctan: {cordic_result}, Error: {error_cordic}")
```
